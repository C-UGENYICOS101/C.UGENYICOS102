# Connecting to a Database

```python
import psycopg2

# Database connection parameters
host = "localhost"
port = "5432"
dbname = "cos101_db"
user = "postgres" # Fixed typo
password = "cos101"

# Establish the connection
try:
    # Changed variable name to 'connection' to match the rest of the script
    connection = psycopg2.connect(
        host=host,
        port=port,
        dbname=dbname,
        user=user,
        password=password
    )
    print("Connection to PostgreSQL DB successful")

    # Create a cursor object
    cursor = connection.cursor()

    # Execute a SQL query - fixed version() syntax
    cursor.execute("SELECT version();")
    
    # Fetch and print the result
    db_version = cursor.fetchone()
    print(f"PostgreSQL database version: {db_version[0]}")

    # Close the cursor and connection
    cursor.close()
    connection.close()
    print("PostgreSQL connection is closed")

except Exception as error:
    print(f"Error connecting to PostgreSQL DB: {error}")
```

# Selecting existing Database

```python
import psycopg2

# Database connection parameters
host = "localhost"
port = "5432"
dbname = "cos101_db"
user = "postgres"
password = "cos101"

# Establish the connection
try:
    connection = psycopg2.connect(
        host=host,
        port=port,
        dbname=dbname,
        user=user,
        password=password
    )
    print("Connection to PostgreSQL DB successful")

    # Create a cursor object
    cursor = connection.cursor()

    # Define the SQL SELECT query
    select_query = "SELECT  FROM company;"

    # Execute the query
    cursor.execute(select_query)

    # Fetch all rows from the executed query
    data = cursor.fetchall()

    print("column", " | ", "column2")
    print("------------------------")
        
    # Iterate through the rows and print them
    for row in data:
        print(f"{row[0]}       |  {row[1]}")

    # Close the cursor and connection
    cursor.close()
    connection.close()

except Exception as error:
    print(f"Error connecting to PostgreSQL DB: {error}")
```

# Create a new Database

```python
import psycopg2
from psycopg2 import sql  # Crucial for dynamic SQL identifiers

# Database connection parameters
host = "localhost"
port = "5432"
dbname = "cos101_db"
user = "postgres"
password = "cos101"

# Fixed the variable name typo here
new_dbname = "cos102_db" 

try:
    # Connect to the PostgreSQL server
    connection = psycopg2.connect(
        host=host,
        port=port,
        dbname=dbname,
        user=user,
        password=password
    )
    
    # Databases cannot be created inside a transaction block
    connection.autocommit = True 
    print("Connection to PostgreSQL server successful")

    # Create a cursor object
    cursor = connection.cursor()

    # Define the SQL statement safely
    # This prevents SQL injection by wrapping the name in double quotes
    create_db_query = sql.SQL("CREATE DATABASE {}").format(sql.Identifier(new_dbname))

    # Execute the SQL statement
    cursor.execute(create_db_query)
    print(f"Database '{new_dbname}' created successfully")

    # Close the cursor and connection
    cursor.close()
    connection.close()

except Exception as error:
    print(f"Error creating the database: {error}")
```

# Create a Table

```python
import psycopg2

# Database connection parameters
host = "localhost"
port = "5432"
dbname = "cos102_db"
user = "postgres"
password = "cos101"

try:
    # Connect to the PostgreSQL database
    connection = psycopg2.connect(
        host=host,
        port=port,
        dbname=dbname,
        user=user,
        password=password
    )
    
    print("Connection to PostgreSQL DB successful")

    # Create a cursor object (fixed name from 'cursors' to 'cursor')
    cursor = connection.cursor()

    # Define the SQL statement - removed the extra comma after the last column
    create_table_query = """
    CREATE TABLE sst_info (
        id SERIAL PRIMARY KEY,
        name VARCHAR(50) NOT NULL,
        age INT NOT NULL,
        department VARCHAR NOT NULL,
        password VARCHAR NOT NULL
    );
    """

    # Execute the SQL statement
    cursor.execute(create_table_query)
    
    # Commit the transaction to save changes
    connection.commit()
    print("Table created successfully")

    # Close the cursor and connection
    cursor.close()
    connection.close()

except Exception as error:
    print(f"Error creating table: {error}")
```

# INSERT into Table

```python
import psycopg2

# Database connection parameters
host = "localhost"
port = "5432"
dbname = "cos102_db"
user = "postgres"
password = "cos101"

try:
    connection = psycopg2.connect(
        host=host,
        port=port,
        dbname=dbname,
        user=user,
        password=password
    )
    connection.autocommit = True
    print("Connection to PostgreSQL DB successful")

    cursor = connection.cursor()

    # Fixed: Removed 'id' from the column list and values
    # PostgreSQL will auto-generate the ID because it is a SERIAL type
    insert_query = """
    INSERT INTO sst_info (name, age, department, password)
    VALUES (%s, %s, %s, %s);
    """
    
    # Data to be inserted (matching the 4 placeholders)
    data_to_insert = ("Jamel Ojo", 20, "Software Engineering", "anthelo4")

    # Execute the SQL statement
    cursor.execute(insert_query, data_to_insert)
    print("Data inserted successfully")

    # Close the cursor and connection
    cursor.close()
    connection.close()

except Exception as error:
    print(f"Error inserting data: {error}")
```

# SELECT from Table

```python
import psycopg2

# Database connection parameters
db_params = {
    'dbname': "cos102_db",
    'user': "postgres",
    'password': "cos101",
    'host': "localhost",
    'port': "5432"
}

# Initialize variables as None for the 'finally' block safety
connection = None
cursor = None

try:
    connection = psycopg2.connect(**db_params)
    cursor = connection.cursor()

    # Bug Fix 1: Added '*' to actually select the data columns
    select_query = "SELECT * FROM sst_info WHERE age < 30"

    cursor.execute(select_query)

    # Fetch all the results
    results = cursor.fetchall()

    # Bug Fix 2: Changed 'result' to 'results' to match the variable above
    if not results:
        print("No records found matching that criteria.")
    else:
        for row in results:
            print(row)

except Exception as error:
    print(f"Error while connecting to PostgreSQL: {error}")

finally:
    # Bug Fix 3: Standard safety check for closing
    if cursor:
        cursor.close()
    if connection:
        connection.close()
    print("PostgreSQL connection closed.")
```
