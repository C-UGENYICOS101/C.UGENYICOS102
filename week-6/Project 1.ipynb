```python
import tkinter as tk
from tkinter import messagebox
import random
from datetime import datetime

class Employee:
    def __init__(self):
        self.staff_list = [
            "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", 
            "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James", 
            "Michell Taiwo", "Abraham Jones", "Nicole Anide", "Kosi Korso", 
            "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
        ]
        self.tasks = ["Loading", "Transporting", "Reviewing Orders", "Customer Service", "Delivering Items"]

    def check_employee(self, name):
        return any(staff.lower() == name.lower() for staff in self.staff_list)

    def take_attendance(self, name):
        now = datetime.now().strftime("%H:%M:%S")
        return f"Attendance logged for {name} at {now}."

    def assign_task(self):
        index = random.randint(0, len(self.tasks) - 1)
        return self.tasks[index]

    def refuse_access(self):
        return "Access Denied. You are not recognized in our employee database."

def process_login():
    name = entry_name.get().strip()
    system = Employee()

    if system.check_employee(name):
        attendance_msg = system.take_attendance(name)
        task = system.assign_task()
        
        result_text = f"Welcome, {name}!\n\n{attendance_msg}\nYour task for today is: {task}"
        messagebox.showinfo("Success", result_text)
    else:
        messagebox.showerror("Unauthorized", system.refuse_access())

root = tk.Tk()
root.title("Mrs. Jane's Logistics System")
root.geometry("400x250")

label_title = tk.Label(root, text="Employee Biometric Login", font=("Arial", 14, "bold"))
label_title.pack(pady=20)

label_prompt = tk.Label(root, text="Enter Full Name:")
label_prompt.pack()

entry_name = tk.Entry(root, width=30)
entry_name.pack(pady=5)

btn_login = tk.Button(root, text="Verify & Log In", command=process_login, bg="green", fg="white")
btn_login.pack(pady=20)

root.mainloop()
```
