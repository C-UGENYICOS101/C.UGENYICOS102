```python
import tkinter as tk
from tkinter import messagebox

class DeliveryService:
    def __init__(self):
        self.prices = {
            "PAU": {"heavy": 2000, "light": 1500},
            "Epe": {"heavy": 5000, "light": 4000}
        }

    def calculate_fee(self, location, weight):
        if weight >= 10:
            return self.prices[location]["heavy"]
        else:
            return self.prices[location]["light"]

def process_payment():
    try:
        weight = float(entry_weight.get())
        location = location_var.get()
        
        service = DeliveryService()
        fee = service.calculate_fee(location, weight)
        
        messagebox.showinfo("Quote", f"Delivery to {location}\nWeight: {weight}kg\nTotal Charge: N{fee}")
    except ValueError:
        messagebox.showerror("Error", "Please enter a valid number for weight.")

root = tk.Tk()
root.title("Delivery Pricing System")
root.geometry("350x300")

tk.Label(root, text="Package Delivery Quote", font=("Arial", 12, "bold")).pack(pady=10)

tk.Label(root, text="Select Location:").pack()
location_var = tk.StringVar(value="PAU")
tk.Radiobutton(root, text="PAU", variable=location_var, value="PAU").pack()
tk.Radiobutton(root, text="Epe", variable=location_var, value="Epe").pack()

tk.Label(root, text="Enter Weight (kg):").pack(pady=(10, 0))
entry_weight = tk.Entry(root)
entry_weight.pack()

btn_calculate = tk.Button(root, text="Calculate Fee", command=process_payment, bg="blue", fg="white")
btn_calculate.pack(pady=20)

root.mainloop()
```
